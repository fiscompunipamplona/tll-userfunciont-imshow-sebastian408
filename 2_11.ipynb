```
def comb(n1,n2):
  from math import factorial as fact
  i=n1;M1=1
  while i>n2:
    M1=i*M1
    i-=1
  M2=M1/fact(n1-n2)
  return M2

for i in range(1,21):
  j=0;RTA=[]
  for j in range(0,i+1):
    RTA.append(round(comb(i, j),1))
  print(RTA,"\n")
```
